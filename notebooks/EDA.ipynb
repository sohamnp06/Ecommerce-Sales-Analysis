## Customer Review & Feedback Summary

The following observations were made from the customer review data:

- Customers rated products on a **5-point review rating scale**, ranging from **1 (lowest)** to **5 (highest)**.
- The review ratings were categorized into **three feedback groups**: **Positive**, **Neutral**, and **Negative**.
- A total of **6,150 reviews** were classified as **Positive**, **3,013 reviews** as **Neutral**, and **5,837 reviews** as **Negative**.
- Each feedback category consists of meaningful customer review texts, allowing customer opinions to be analyzed more effectively.
- Positive feedback includes comments such as *"Excellent product!"*, *"Highly recommend!"*, and *"Value for money"*, while negative feedback highlights issues like *"Poor packaging"*, *"Waste of money"*, and *"Defective item"*.

### Conclusion

The customer reviews were successfully transformed into three meaningful feedback categories—**Positive**, **Neutral**, and **Negative**. This sentiment grouping simplifies customer opinion analysis and provides a strong foundation for identifying satisfaction trends, evaluating product performance, and generating actionable business insights.
