The following observations were made from the customer review data:

### Conclusion

The customer reviews were successfully transformed into three meaningful feedback categories—**Positive**, **Neutral**, and **Negative**. This sentiment grouping simplifies customer opinion analysis and provides a strong foundation for identifying satisfaction trends, evaluating product performance, and generating actionable business insights.

# Business Insights

## 1. State-wise Returned Orders

- **Assam (201)** recorded the highest number of returned orders, followed by **Uttar Pradesh (196)** and **Sikkim (193)**.
- **Andhra Pradesh (139)** had the lowest number of returned orders among all states.
- The difference in return counts across states is relatively small, indicating that product returns are fairly consistent throughout the regions.

**Business Insight:**  
States with higher return volumes, particularly Assam and Uttar Pradesh, should be investigated to identify whether returns are driven by product quality, delivery issues, incorrect orders, or customer expectations.

---

## 2. State-wise Customer Feedback Analysis

### States with the Highest Positive Feedback

- **Jharkhand** recorded the highest percentage of positive customer feedback (**45.16%**).
- **Gujarat (44.05%)**, **Andhra Pradesh (43.96%)**, **Bihar (43.90%)**, and **Tripura (43.89%)** also demonstrated high customer satisfaction.
- These states consistently showed higher customer satisfaction compared to the overall average.

### States with the Highest Negative Feedback

- **Chhattisgarh** recorded the highest percentage of negative feedback (**44.06%**).
- Other states with relatively higher negative feedback include **Arunachal Pradesh**, **Goa**, **Telangana**, and **Meghalaya**.

**Business Insight:**  
Customer satisfaction varies across states. States with higher negative feedback should be prioritized for quality improvement initiatives, customer service enhancements, and operational reviews to improve the overall customer experience.
